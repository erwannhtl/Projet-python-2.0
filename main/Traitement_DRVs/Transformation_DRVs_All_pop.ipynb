```python
import pandas as pd
import re

# Charger le fichier CSV dans un DataFrame pandas
df = pd.read_csv('Traitement_DRVs/DRVs_All_populations_clean.csv')

# Définir une fonction pour extraire la catégorie d'âge et le PAL
def extract_age_and_pal(age_string):
    match = re.search(r'(≥?\s?\d+[-–]?\d*|\d+[-–]?\d*)\s?(?:years?)?(?: PAL=([0-9.]+))?', age_string)
    if match:
        years = match.group(1)
        # Traitement spécifique pour '≥ 18 years'
        if '≥ 18' in years:
            years = '≥ 18'
        pal = float(match.group(2)) if match.group(2) is not None else None
        return years, pal
    else:
        return None, None

# Appliquer la fonction à la colonne "Age" pour créer une nouvelle colonne "Years" et "PAL"
df['Years'], df['PAL'] = zip(*df['Age'].apply(extract_age_and_pal))

import pandas as pd

# Définir les remplacements à effectuer
remplacements = {
    'Pregnant women (1st trimester)': 'Pregnant women',
    'Pregnant women (2nd trimester)': 'Pregnant women',
    'Pregnant women (3rd trimester)': 'Pregnant women',
    'Lactating women (0–6 months post partum)': 'Lactating women',
    'Lactating women (>6 months post partum)': 'Lactating women'
}

# Appliquer les remplacements à la colonne 'Target population'
df['Target population'] = df['Target population'].replace(remplacements)

# Afficher les premières lignes du DataFrame résultant
print(df[['Age', 'Years', 'PAL']].head())

# Sauvegarder le DataFrame modifié au format CSV
df.to_csv('Traitement_DRVs/nouveau_fichier_DRVs.csv', index=False)
```

```python
import pandas as pd

# Charger votre DataFrame à partir d'un fichier CSV
# Assurez-vous d'ajuster le chemin du fichier CSV selon votre environnement
df = pd.read_csv('Traitement_DRVs/nouveau_fichier_DRVs.csv')

# Identifier les lignes correspondant aux critères spécifiés
criteria1 = (df['Target population'] == 'Pregnant women') & (df['Nutrient'] == 'Energy')
criteria2 = (df['Target population'] == 'Pregnant women') & (df['Nutrient'] == 'Protein')
criteria3 = (df['Target population'] == 'Pregnant women') & (df['Nutrient'] == 'Calcium')
criteria4 = (df['Target population'] == 'Lactating women') & (df['Nutrient'] == 'Energy')
criteria5 = (df['Target population'] == 'Lactating women') & (df['Nutrient'] == 'Protein')
criteria6 = (df['Target population'] == 'Lactating women') & (df['Nutrient'] == 'Calcium')

# Combiner toutes les conditions
mask = criteria1 | criteria2 | criteria3 | criteria4 | criteria5 | criteria6

# Sélectionner la première ligne pour chaque ensemble de critères spécifiés
filtered_df = df[mask].drop_duplicates(subset=['Target population', 'Nutrient'])

# Inclure toutes les autres lignes qui ne répondent à aucun de ces critères
other_rows = df[~mask]

# Concaténer les deux résultats
final_df = pd.concat([filtered_df, other_rows])
print(final_df)
```

```python
import pandas as pd

def convert_ai_to_kcal(value):
    if pd.notna(value):
        # Supprimer le '(+)' s'il est présent
        value = value.replace('(+) ', '')

        # Extraire la valeur numérique et l'unité
        match = re.match(r'([\d.]+)\s*([a-zA-Z/]+)', value)
        if match:
            numeric_value, unit = match.groups()

            # Convertir en MJ et puis en kcal
            if unit.lower() == 'mj/day':
                mj_value = float(numeric_value)
                kcal_value = mj_value * 238.83
                return kcal_value

    # Retourner la valeur inchangée si la conversion n'est pas possible
    return value
    
# Appliquer la conversion à la colonne AI si la colonne Nutrient est 'Energy'
final_df.loc[final_df['Nutrient'] == 'Energy', 'AR'] = final_df.loc[final_df['Nutrient'] == 'Energy', 'AR'].apply(convert_ai_to_kcal)
```

```python
import pandas as pd
import numpy as np

# Exemple de DataFrame
data = final_df

df = pd.DataFrame(data)

def convert_ai(column):
    # Si le nutrient est "Water", supprimer " L/day" et multiplier par 1000
    if column['Nutrient'] == 'Water':
        return float(column['AI'].replace(' L/day', '')) * 1000
    # Si le nutrient est l'un de ceux spécifiés, extraire la valeur numérique
    elif column['Nutrient'] in ['Magnesium', 'Manganese', 'Potassium', 'Vitamin E', 'Vitamin D']:
        return pd.to_numeric(column['AI'].split(' ')[0], errors='coerce')
    # Sinon, conserver la valeur telle quelle
    else:
        return column['AI']

# Appliquer la fonction à la colonne "AI"
df['AI'] = df.apply(convert_ai, axis=1)

df_presque_fin = df.copy()
```

```python
import pandas as pd

# Exemple de DataFrame
data = df_presque_fin
df = pd.DataFrame(data)

def convert_ar(row):
    # Si le nutrient est "Protein", supprimer le "+" et ne garder que le float
    if row['Nutrient'] == 'Protein':
        ar_value = str(row['AR']).replace('(+', '').replace(')', '').split(' ')[0]
        return float(ar_value) if ar_value.lower() != 'na' and ar_value else None
    # Si le nutrient est l'un de ceux spécifiés, ne garder que le float
    elif row['Nutrient'] in ['Calcium', 'Vitamin B6', 'Vitamin C']:
        ar_value = str(row['AR']).split(' ')[0]
        return float(ar_value) if ar_value.lower() != 'na' and ar_value else None
    # Sinon, conserver la valeur telle quelle
    else:
        try:
            return float(row['AR'])
        except ValueError:
            return None

# Appliquer la fonction à la colonne "AR"
df['AR'] = df.apply(convert_ar, axis=1)

# Nommer le DataFrame modifié
df_derniere = df.copy()

# Afficher le DataFrame résultant
print(df_derniere)
```

```python
import pandas as pd

# Charger le DataFrame original depuis le fichier CSV
data = df_derniere
df = pd.DataFrame(data)

def convert_sai(row):
    # Si le nutrient est "Sodium", extraire la valeur numérique et multiplier par 1000
    if row['Nutrient'] == 'Sodium':
        sai_value = str(row['Safe and adequate intake'])
        try:
            if 'g/day' in sai_value:
                return float(sai_value.replace(' g/day', '')) * 1000
            else:
                return float(sai_value)
        except ValueError:
            return None
    # Sinon, conserver la valeur telle quelle
    else:
        try:
            return float(row['Safe and adequate intake'])
        except ValueError:
            return None

# Appliquer la fonction à la colonne "Safe and adequate intake"
df['Safe and adequate intake'] = df.apply(convert_sai, axis=1)

# Nommer le DataFrame modifié
df_derniere = df.copy()

# Afficher le DataFrame résultant
df.to_csv('Traitement_DRVs/final_DRVs.csv', index=False)
```
