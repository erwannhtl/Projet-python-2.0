# Webscrapping

Nous avons décidé de webscrapper le site https://1ou2cocktails.com/cocktails/ car il nous semblait que c'était un des sites les plus exhaustifs en français. 

Le problème principal lors du webscrapping était que les informations sur les ingrédients n'étaient pas accesibles dans le html : la page dynamique est codée en partie en Javascript en nous ne pouvions pas récupérer les données avec BeautifulSoup ou autre. 
Nous avons donc directement demandé au serveur de nous afficher les 410 hitperpage au lieu des 40 normalement affiché en récupérant la clef API.
Grâce à cette commande, nous avons récupéré un fichier json avec toutes les informations sur les cocktails.

```python
import requests

num_cocktails = 410

json = {"requests":[{"indexName":"production_content","params":"query=&maxValuesPerFacet=100&page=0&highlightPreTag=__ais-highlight__&highlightPostTag=__%2Fais-highlight__&filters=post_type%3Acocktail%20AND%20locale%3Afr&hitsPerPage="+str(num_cocktails)+"&facets=%5B%22spirit%22%2C%22type%22%2C%22taste%22%2C%22appearance%22%2C%22thematic%22%2C%22occasion%22%5D&tagFilters="}]}
params = {"x-algolia-api-key" : "a0e3904858756519cc762015907535d8", "x-algolia-application-id" : "OK8ZSR9FTP"}

r = requests.post("https://ok8zsr9ftp-dsn.algolia.net/1/indexes/*/queries?x-algolia-agent=Algolia%20for%20JavaScript%20(4.8.4)%3B%20Browser%3B%20instantsearch.js%20(4.21.0)%3B%20JS%20Helper%20(3.4.4)", json=json, params=params)

if r.status_code != 200:
    print("Il faut changer la clef API")

print(r.text)
```
